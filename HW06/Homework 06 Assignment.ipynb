#### Question 1 - Read in the data set and print the % empty values for each column. Are there any columns with missing values?

```python
data_dir = '/Users/ananyasingh/CS301-DataScience/HW06/groundhog.csv'
dataframe = pd.read_csv(data_dir)
```

```python
def get_percent_nans(df, column_name):
    specific_column = df[column_name]
    return (len(df.loc[specific_column.isna()])/len(df))*100
```

```python
for col in dataframe.columns:
    print('The {} column has {:.2f}% empty values'.format(col,get_percent_nans(dataframe,col)))
```

#### Question 6 - For the remaining rows which have no record (but the other columns *are* populated), determine an appropriate method for inputting the value for this column. Explain your reasoning, and show your work below (using pandas).

```python
nRFebMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Record')]['February Average Temperature'].mean()
nRMarMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Record')]['March Average Temperature'].mean()
nRAvg = (nRFebMean+nRMarMean)/2
print('The average temperature for NO RECORD is : ' + str(nRAvg))
```

```python
nSFebMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Shadow')]['February Average Temperature'].mean()
nSMarMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Shadow')]['March Average Temperature'].mean()
nSAvg = (nSFebMean+nSMarMean)/2
print('The average temperature for NO SHADOW is : ' + str(nSAvg))
```

```python
pSFebMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'Partial Shadow')]['February Average Temperature'].mean()
pSMarMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'Partial Shadow')]['March Average Temperature'].mean()
pSAvg = (pSFebMean+pSMarMean)/2
print('The average temperature for PARTIAL SHADOW is : ' + str(pSAvg))
```

```python
fSFebMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'Full Shadow')]['February Average Temperature'].mean()
fSMarMean = dataframe.loc[(dataframe['Punxsutawney Phil'] == 'Full Shadow')]['March Average Temperature'].mean()
fSAvg = (fSFebMean+fSMarMean)/2
print('The average temperature for FULL SHADOW is : ' + str(fSAvg))
```

```python
noShadowDifference = nSAvg - nRAvg
print("The difference between No Shadow Average and No Record Average is: " + str(noShadowDifference))
partialShadowDifference = pSAvg - nRAvg
print("The difference between Partial Shadow Average and No Record Average is: " + str(partialShadowDifference))
fullShadowDifference = fSAvg - nRAvg
print("The difference between Full Shadow Average and No Record Average is: " + str(fullShadowDifference))
```

The difference between the overall Partial Shadow average and overall No Record Temperature average is the least compared to other shadows, therefore I will replace all No Records with Partial Shadow. I don't think it will have a big effect overall. 

```python
dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Record'), 'Shadow'] = 1
dataframe.loc[(dataframe['Punxsutawney Phil'] == 'No Record'), 'Punxsutawney Phil'] = 'Partial Shadow'
```
